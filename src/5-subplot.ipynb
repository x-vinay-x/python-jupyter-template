### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Data for the line chart
x_line = np.arange(0, 10)
y_line = x_line ** 2

# Plot a line chart in the first subplot
axs[0, 0].plot(x_line, y_line, color='blue', marker='o')
axs[0, 0].set_title('Line Chart (y = x^2)')
axs[0, 0].set_xlabel('x values')
axs[0, 0].set_ylabel('y values')

# Data for the bar chart
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Plot a bar chart in the second subplot
axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Data for the scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Plot a scatter plot in the third subplot
axs[1, 0].scatter(x_scatter, y_scatter, color='purple')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('x values')
axs[1, 0].set_ylabel('y values')

# Data for the pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Plot a pie chart in the fourth subplot
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'], startangle=140)
axs[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```
